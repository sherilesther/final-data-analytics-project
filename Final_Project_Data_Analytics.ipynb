```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_sales_data.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
# Fill numeric missing values
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df.describe()
```

```python
df["Sales"].hist()
plt.title("Sales Distribution")
plt.show()
```

```python
df.groupby("Category")["Sales"].sum().plot(kind="bar")
plt.title("Total Sales by Category")
plt.show()
```

```python
sns.boxplot(x=df["Sales"])
plt.title("Sales Outliers")
plt.show()
```

```python
plt.scatter(df["Quantity"], df["Sales"])
plt.xlabel("Quantity")
plt.ylabel("Sales")
plt.title("Quantity vs Sales")
plt.show()
```

## Insights & Observations

- Sales values are mostly concentrated around the average range.
- A few high sales values appear as outliers, possibly bulk purchases.
- Some categories contribute significantly more to total sales.
- Quantity and sales show a positive relationship.
